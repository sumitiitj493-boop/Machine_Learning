```python
import numpy as np
import pandas as pd
```

 CSV file ko DataFrame mein load kiya aur pehle 5 rows ka preview

###4. Step 3: Feature Selection

Yaha dono columns (IQ, CGPA) important hain. Sab use karenge.
Real world mein bahut saare columns hote hain, toh decide karna padta hai kaunse lene hain.

###5. Step 4: Input-Output Columns Extract Karna

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
x.shape
```

2D tensor (matrix)

```python
y.shape
```

1D tensor (vector)

###6. Step 5: Train-Test Split

Agar 100% data pe train kar diya, toh evaluate kaise karoge?

Solution: Kuch data chhupa ke rakho (test set). Baaki pe train karo (training set). Phir
chhupaye data pe test karke dekho model kitna sahi hai.

•90 students ka data Training Set (model seekhega isse)

• 10 students ka data Test Set (model ko evaluate karenge isse)

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
x_train.shape # (90, 2) 90 students, 2 features
```

```python
x_test.shape # (10, 2) 10 students, 2 features
```

Randomly rows select hote hain — isliye har baar run karne pe alag rows training/test mein
ja sakte hain.

###7. Step 6: Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

 # Training data pe fit + transform
x_train = scaler.fit_transform(x_train)

 # Test data pe sirf transform (fit nahi!)
x_test = scaler.transform(x_test)
```

**fit transform vs transform**

• **fit transform** pehle pattern seekho (fit), phir apply karo (transform)

• **transform** sirf apply karo (pattern pehle hi seekh liya training se)

###8. Step 7: Model Training

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression() # classifier object banaya

clf.fit(x_train, y_train) # BOOM! Model trained!
```

###9. Step 8: Model Evaluation

9.1 Take Predictions

```python
y_pred = clf.predict(x_test) # test data pe predictions

print(y_pred) # model ke predictions
print(y_test) # actual values (ground truth)
```

9.2 Accuracy Calculation

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(accuracy) # 1 = 100% accuracy
```

Accuracy = Correct Predictions/Total Predictions

9.3 Decision Boundary Visualization

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x_train, y_train.values, clf=clf)
plt.show()
```

###10. Step 9: Model Export Karna (Pickle)

Model ready hai, ab isko save karna hai taaki baad mein use kar sakein.

```python
import pickle
# Model ko file mein save karo
pickle.dump(clf, open('model.pkl', 'wb'))
# ’wb’ = write binary mode
```

Ab model.pkl file bani — isko website mein use kar sakte hain.

**Pickle kya karta hai?**

Python ka koi bhi object le ke usko file mein convert kar deta hai. Phir us file ko kisi bhi
