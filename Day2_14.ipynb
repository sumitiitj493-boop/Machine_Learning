## Case Study: Netflix Churn Rate Problem

**Churn Rate Kya Hai?**

Churn Rate = Kitne percent customers ek given period (monthly/yearly) mein platform
chhod ke jaate hain.

The Setup

**Netflix ka revenue kaise badhaye?**

1. Naye customers lao
2. Existing customers se zyada charge karo
3. Jo customers jaane wale hain, unko roko


###Step 1: Business Mathematical Problem

Jab bhi business problem mile, pehla kaam hai usse ek measurable mathematical goal
mein convert karna.

**Before**:"Revenue badhao" (vague)

**After**: "Churn rate 4% se 3.75% pe laao in 6 months" (specific, measurable)

###Step 2: Identify the Type of Problem

**2.1** *Goal Clear Karo*

Identify Risky Customers ------>
Take Action to Retain--------->
Churn Rate Reduces

 **2.2** *Rokne ke Liye Kya Karoge?*

1. **Immediate Action:** Next month ka subscription 50% discount pe de do

2. **Root Cause Analysis:**

• Pasand ka content nahi mil raha?

• UI navigate nahi kar pa raha (old age user)?

• Internet problem hai uske area mein?

**2.3** *Classification ya Regression?*

**Pehle socha:** Har customer ke liye predict karo — chhod ke jaayega ya nahi (Yes/No).
Ye toh Classification Problem ho gaya!

**Better Thinking :**

Sab customers ko equal kyun maan rahe hain?

• Kisi ka bahut zyada mann hai chhod ke jaane ka

• Kisi ka thoda bahut mann hai chhod ke jaane ka

**Final Decision: Regression Problem**

Output =Har customer ke liye probability score (0-100%) ki wo platform chhod ke jaayega.
Jiska score jitna high Usko utna zyada discount/attention do.

###Step 6: Online Learning vs Batch Learning

• Lockdown laga Zyada log aaye, zyada time spend kiya

• Lockdown hata Bahut log chale gaye (jo sirf time pass ke liye aaye the)
